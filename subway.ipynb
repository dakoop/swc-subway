```python
# functions have a name followed parenthesis with
# arguments in the middle
def read_subway_data(filename):
    """read_subway_data(filename: string) -> dictionary
    of stations and fares
    """
    f = open(filename, "r")
    reader = csv.reader(f)
    
    clean_data2 = {}
    # skip the first 3 lines
    r_iter = iter(reader)
    for i in xrange(3):
        r_iter.next()
    for row in r_iter:
        station = row[1].strip()
        fares = []
        for fare in row[2:]:
            if fare.strip() != "":
                fares.append(int(fare))
        clean_data2[station] = fares

    # send back the data to whoever called this function
    return clean_data2
```
